### PadronizaÃ§Ã£o dos nomes das classes

Como o dataset pode conter pastas com nomes inconsistentes (ex.: `stawberries`, `pinenapple`), esta etapa garante que **todas as classes tenham nomes padronizados e corrigidos** antes do treinamento.

O cÃ³digo faz:

1. **LÃª automaticamente** os nomes das pastas dentro dos diretÃ³rios `train` e `test`;
2. Aplica um dicionÃ¡rio de correÃ§Ãµes (`standard_names`), corrigindo erros de digitaÃ§Ã£o e padronizando capitalizaÃ§Ã£o;
3. Retorna a lista final ordenada das classes jÃ¡ corrigidas.

Isso evita que classes diferentes sejam tratadas como distintas apenas devido a erros de escrita no nome da pasta.


```python
import os

# DicionÃ¡rio de padronizaÃ§Ã£o
standard_names = {
    'apple': 'Apple',
    'banana': 'Banana',
    'avocado': 'Avocado',
    'cherry': 'Cherry',
    'kiwi': 'Kiwi',
    'mango': 'Mango',
    'orange': 'Orange',
    'pinenapple': 'Pineapple',
    'pineapple': 'Pineapple',
    'strawberries': 'Strawberries',
    'stawberries': 'Strawberries',
    'watermelon': 'Watermelon'
}

def standardize_class_names(class_list):
    return sorted([standard_names.get(c.lower(), c) for c in class_list])

# âš™ï¸ LÃª as classes diretamente das pastas
train_dir = '/home/wallingson12/Github/Data science and ML/Classificador de frutas/train'
validation_dir = '/home/wallingson12/Github/Data science and ML/Classificador de frutas/test'

train_classes_raw = [d for d in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, d))]
val_classes_raw = [d for d in os.listdir(validation_dir) if os.path.isdir(os.path.join(validation_dir, d))]

# âœ… Padroniza
train_std = standardize_class_names(train_classes_raw)
val_std   = standardize_class_names(val_classes_raw)
```

### VerificaÃ§Ã£o das classes apÃ³s padronizaÃ§Ã£o

ApÃ³s corrigir e padronizar os nomes das pastas, Ã© importante confirmar se o conjunto de **treino** e o de **validaÃ§Ã£o** possuem exatamente as mesmas classes.

O cÃ³digo:

- Exibe no console as classes padronizadas de cada conjunto;
- Compara ambas as listas para garantir que nÃ£o existe uma fruta presente somente em um dos conjuntos;
- Caso haja discrepÃ¢ncias, ele indica quais classes estÃ£o faltando em cada conjunto.

Isso evita erros futuros, como a rede ser treinada com uma classe que nÃ£o aparece no conjunto de validaÃ§Ã£o, o que afetaria as mÃ©tricas de avaliaÃ§Ã£o e o processo de generalizaÃ§Ã£o.

```python
print("â–¶ Classes padronizadas (treino):", train_std)
print("â–¶ Classes padronizadas (validaÃ§Ã£o):", val_std)

# Verifica se hÃ¡ classes faltando em algum conjunto
missing_in_val = set(train_std) - set(val_std)
missing_in_train = set(val_std) - set(train_std)

if not missing_in_val and not missing_in_train:
    print("\nPerfeito! Treino e validaÃ§Ã£o agora possuem as MESMAS classes.")
else:
    print("\nâš  DiferenÃ§as encontradas:")
    if missing_in_val:
        print(" - Presentes no treino mas ausentes na validaÃ§Ã£o:", missing_in_val)
    if missing_in_train:
        print(" - Presentes na validaÃ§Ã£o mas ausentes no treino:", missing_in_train)
```

**Data Augmentation e PrÃ©-processamento**

```python
# BASE
from tensorflow.keras import layers, models, regularizers
import tensorflow as tf

def build_model(image_size, num_classes):
    data_augmentation = tf.keras.Sequential([
        layers.RandomFlip("horizontal"),
        layers.RandomRotation(0.10),
        layers.RandomZoom(0.20),
        layers.RandomContrast(0.20),
        layers.RandomBrightness(0.20),
        layers.RandomTranslation(0.10, 0.10),
    ])

    l2 = regularizers.l2(1e-4)

    model = models.Sequential([
        layers.Input(shape=(image_size[0], image_size[1], 3)),
        data_augmentation,
        layers.Rescaling(1./255),

        # Bloco 1
        layers.Conv2D(32, 3, padding='same', activation='relu', kernel_regularizer=l2),
        layers.BatchNormalization(),
        layers.Conv2D(32, 3, padding='same', activation='relu', kernel_regularizer=l2),
        layers.BatchNormalization(),
        layers.MaxPooling2D(),
        layers.Dropout(0.15),

        # Bloco 2
        layers.Conv2D(64, 3, padding='same', activation='relu', kernel_regularizer=l2),
        layers.BatchNormalization(),
        layers.Conv2D(64, 3, padding='same', activation='relu', kernel_regularizer=l2),
        layers.BatchNormalization(),
        layers.MaxPooling2D(),
        layers.Dropout(0.20),

        # Bloco 3
        layers.Conv2D(128, 3, padding='same', activation='relu', kernel_regularizer=l2),
        layers.BatchNormalization(),
        layers.Conv2D(128, 3, padding='same', activation='relu', kernel_regularizer=l2),
        layers.BatchNormalization(),
        layers.MaxPooling2D(),
        layers.Dropout(0.25),

        layers.GlobalAveragePooling2D(),

        layers.Dense(256, activation='relu', kernel_regularizer=l2),
        layers.Dropout(0.35),
        layers.Dense(num_classes, activation='softmax')
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-4),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
```

**Treinar**

```python
import os
import tensorflow as tf
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint
import pandas as pd

def train_model(model, train_dir, validation_dir, image_size=(224,224), batch_size=32, epochs=50, save_dir="./histories"):
    """
    Treina o modelo e salva o melhor modelo + histÃ³rico.
    """
    os.makedirs(save_dir, exist_ok=True)

    print("Carregando datasets...")
    train_dataset = tf.keras.utils.image_dataset_from_directory(
        train_dir,
        image_size=image_size,
        batch_size=batch_size,
        label_mode='categorical',
        shuffle=True
    )
    val_dataset = tf.keras.utils.image_dataset_from_directory(
        validation_dir,
        image_size=image_size,
        batch_size=batch_size,
        label_mode='categorical',
        shuffle=False
    )

    # Callbacks
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1, min_lr=1e-6)
    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)
    checkpoint = ModelCheckpoint(filepath=os.path.join(save_dir, 'best_model.keras'), monitor='val_accuracy', save_best_only=True, verbose=1)

    print("Iniciando treino...")
    history = model.fit(
        train_dataset,
        epochs=epochs,
        validation_data=val_dataset,
        callbacks=[reduce_lr, early_stop, checkpoint]
    )

    # Salvar histÃ³rico
    history_file = os.path.join(save_dir, 'history.csv')
    pd.DataFrame(history.history).to_csv(history_file, index=False)
    print(f"HistÃ³rico salvo em: {history_file}")
    print(f"Modelo salvo em: {os.path.join(save_dir, 'best_model.keras')}")

    return history
```

### ConstruÃ§Ã£o do Modelo CNN (Arquitetura Base)

A funÃ§Ã£o `build_model()` cria uma **rede neural convolucional (CNN)** projetada para classificaÃ§Ã£o de imagens de frutas.  
A arquitetura utiliza regularizaÃ§Ã£o, data augmentation e camadas convolucionais empilhadas para melhorar o desempenho e reduzir overfitting.

**Componentes principais:**

- **Data Augmentation**  
  Aplicado diretamente na entrada do modelo para aumentar a variabilidade das imagens:
  - Flip horizontal  
  - RotaÃ§Ã£o  
  - Zoom  
  - Contraste  
  - Brilho  
  - TranslaÃ§Ã£o

- **NormalizaÃ§Ã£o**
  As imagens sÃ£o reescaladas para valores entre `0 e 1` com `layers.Rescaling(1./255)`.

- **Arquitetura Convolucional**
  O modelo possui **3 blocos convolucionais**, cada um contendo:
  - 2 camadas `Conv2D` com ativaÃ§Ã£o ReLU  
  - `BatchNormalization` apÃ³s cada convoluÃ§Ã£o
  - `MaxPooling2D` para reduÃ§Ã£o espacial
  - `Dropout` crescente a cada bloco (0.15 â†’ 0.20 â†’ 0.25)

- **RegularizaÃ§Ã£o L2**
  Aplicada a todas as camadas convolucionais e Ã  camada densa para evitar overfitting.

- **Classificador Final**
  - `GlobalAveragePooling2D` reduz o mapa de caracterÃ­sticas sem achatar totalmente
  - `Dense(256, relu)` com dropout de 0.35
  - `Dense(num_classes, softmax)` para saÃ­da multiclasses

- **CompilaÃ§Ã£o**
  - Otimizador: `Adam (1e-4)`
  - FunÃ§Ã£o de perda: `categorical_crossentropy`
  - MÃ©trica de avaliaÃ§Ã£o: `accuracy`

Essa arquitetura Ã© estÃ¡vel, regularizada e adequada para treinos em datasets com variaÃ§Ã£o visual entre frutas.

```python
import os


# ConfiguraÃ§Ãµes da imagem
image_size = (128, 128)
batch_size = 64
epochs = 40

# Detectando automaticamente as classes pelas pastas do treino
class_names = sorted([d for d in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, d))])
num_classes = len(class_names)

print(f"Classes detectadas: {class_names}")
print(f"NÃºmero de classes: {num_classes}")

# Criando o modelo
model = build_model(image_size=image_size, num_classes=num_classes)

# Treinando o modelo
history = train_model(
    model,
    train_dir,
    validation_dir,
    image_size,
    batch_size=batch_size,
    epochs=epochs,
    save_dir="./histories"
)
```

### Rinha de Deep learning

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import time, os
from sklearn.metrics import classification_report
from tensorflow.keras import layers, models
from rich.console import Console
from rich.table import Table
from rich import box

# ----------------------------
# DiretÃ³rios
# ----------------------------
train_dir = '/home/wallingson12/Github/Data science and ML/Classificador de frutas/train'
val_dir   = '/home/wallingson12/Github/Data science and ML/Classificador de frutas/test'

# ----------------------------
# Modelos Keras Applications
# ----------------------------
MODEL_LIST = {
    "MobileNetV2": tf.keras.applications.MobileNetV2,
    "MobileNetV3Small": tf.keras.applications.MobileNetV3Small,
    "MobileNetV3Large": tf.keras.applications.MobileNetV3Large,
    "ResNet50": tf.keras.applications.ResNet50,
    "ResNet50V2": tf.keras.applications.ResNet50V2,
    "InceptionV3": tf.keras.applications.InceptionV3,
    "Xception": tf.keras.applications.Xception,
    "VGG16": tf.keras.applications.VGG16,
    "VGG19": tf.keras.applications.VGG19,
    "DenseNet121": tf.keras.applications.DenseNet121,
    "DenseNet169": tf.keras.applications.DenseNet169,
    "DenseNet201": tf.keras.applications.DenseNet201,
}

# ----------------------------
# ConfiguraÃ§Ãµes
# ----------------------------
IMG_SIZE = (128, 128)
BATCH = 32
EPOCHS = 8
RESULTS = []

os.makedirs("results", exist_ok=True)

# ----------------------------
# Carregar datasets
# ----------------------------
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    train_dir,
    image_size=IMG_SIZE,
    batch_size=BATCH,
    label_mode="categorical"
)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    val_dir,
    image_size=IMG_SIZE,
    batch_size=BATCH,
    label_mode="categorical"
)

class_names = train_ds.class_names
num_classes = len(class_names)

# ----------------------------
# Console para visualizaÃ§Ã£o
# ----------------------------
console = Console()

# ----------------------------
# Loop de treinamento/teste
# ----------------------------
for name, ModelClass in MODEL_LIST.items():
    console.print(f"\nğŸš€ [bold cyan]Testando modelo: {name}[/bold cyan]")

    try:
        # Base model sem top
        base_model = ModelClass(
            include_top=False,
            weights="imagenet",
            input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3)
        )
        base_model.trainable = False

        # Modelo completo
        model = models.Sequential([
            base_model,
            layers.GlobalAveragePooling2D(),
            layers.Dense(256, activation='relu'),
            layers.Dropout(0.3),
            layers.Dense(num_classes, activation='softmax')
        ])

        model.compile(
            optimizer=tf.keras.optimizers.Adam(1e-4),
            loss="categorical_crossentropy",
            metrics=["accuracy"]
        )

        # Treino
        start = time.time()
        history = model.fit(train_ds, validation_data=val_ds, epochs=EPOCHS, verbose=0)
        train_time = round(time.time() - start, 2)

        # PrevisÃµes para F1
        y_true, y_pred = [], []
        for batch, labels in val_ds:
            preds = model.predict(batch, verbose=0)
            y_true.extend(np.argmax(labels.numpy(), axis=1))
            y_pred.extend(np.argmax(preds, axis=1))

        report = classification_report(y_true, y_pred, output_dict=True)
        f1_macro = report["macro avg"]["f1-score"]

        # Salvar resultado
        RESULTS.append({
            "Modelo": name,
            "AcurÃ¡cia Val (%)": round(report['accuracy'] * 100, 2),
            "F1 Macro": round(f1_macro, 4),
            "Tempo Treino (s)": train_time
        })

        # Salvar histÃ³rico
        pd.DataFrame(history.history).to_csv(f"results/history_{name}.csv", index=False)

        console.print(f"âœ… [green]Finalizado[/green] -> Acc: {report['accuracy']:.4f}, F1: {f1_macro:.4f}, Tempo: {train_time}s")

    except Exception as e:
        console.print(f"âŒ [red]Erro no modelo {name}[/red]: {e}")

# ----------------------------
# Tabela final e ranking
# ----------------------------
df_results = pd.DataFrame(RESULTS)

# Mostrar resultados de cada modelo individualmente
console.print("\nğŸ [bold cyan]Resultados por modelo:[/bold cyan]")
for idx, row in df_results.iterrows():
    table = Table(show_header=True, header_style="bold magenta", box=box.SIMPLE)
    table.add_column("MÃ©trica", style="bold")
    table.add_column("Valor")
    
    table.add_row("Modelo", row['Modelo'])
    table.add_row("AcurÃ¡cia Val (%)", f"{row['AcurÃ¡cia Val (%)']}")
    table.add_row("F1 Macro", f"{row['F1 Macro']}")
    table.add_row("Tempo Treino (s)", f"{row['Tempo Treino (s)']}")
    
    console.print(table)

# Ranking final por F1 Macro
ranking = df_results.sort_values(by='F1 Macro', ascending=False).reset_index(drop=True)
ranking.index += 1  # ComeÃ§a do 1

console.print("\nğŸ† [bold green]Ranking dos modelos (maior F1 Macro primeiro):[/bold green]")

ranking_table = Table(show_header=True, header_style="bold yellow", box=box.DOUBLE_EDGE)
ranking_table.add_column("#", justify="right")
ranking_table.add_column("Modelo")
ranking_table.add_column("AcurÃ¡cia Val (%)", justify="center")
ranking_table.add_column("F1 Macro", justify="center")
ranking_table.add_column("Tempo Treino (s)", justify="center")

for idx, row in ranking.iterrows():
    ranking_table.add_row(
        str(idx + 1),
        row['Modelo'],
        f"{row['AcurÃ¡cia Val (%)']}",
        f"{row['F1 Macro']}",
        f"{row['Tempo Treino (s)']}"
    )

console.print(ranking_table)

# Salvar CSV final
df_results.to_csv("results/metricas_modelos.csv", index=False)
```

### VisualizaÃ§Ã£o do Treinamento â€” AcurÃ¡cia e Loss

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_history(history_csv):
    history = pd.read_csv(history_csv)

    # Plot Accuracy
    plt.figure()
    plt.plot(history['accuracy'])
    plt.plot(history['val_accuracy'])
    plt.title('Treino x ValidaÃ§Ã£o - AcurÃ¡cia')
    plt.xlabel('Ã‰poca')
    plt.ylabel('AcurÃ¡cia')
    plt.legend(['Treino', 'ValidaÃ§Ã£o'])
    plt.show()

    # Plot Loss
    plt.figure()
    plt.plot(history['loss'])
    plt.plot(history['val_loss'])
    plt.title('Treino x ValidaÃ§Ã£o - Loss')
    plt.xlabel('Ã‰poca')
    plt.ylabel('Loss')
    plt.legend(['Treino', 'ValidaÃ§Ã£o'])
    plt.show()

# Exemplo
plot_history("./histories/history.csv")
```
